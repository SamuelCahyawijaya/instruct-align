```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('llm_multi.csv')
```

```python
df.pivot('model', 'lang', 'pct')
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.ticker as mtick

pdf = df.loc[df['model'] != 'mT5 Rescaled'].pivot('model', 'lang', 'pct') * 100
ax = pdf[[
        'English', 'Chinese', 'Code', 'French', 'German',
        'Japanese', 'Portuguese', 'Russian', 'Spanish', 'Others',
    ]].plot(kind='bar', stacked=True, cmap=cm.get_cmap('Spectral'))
plt.legend(loc='upper right', bbox_to_anchor=(1.43, 0.95), fontsize=15)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(rotation=0, fontsize=15)
plt.yticks(rotation=0, fontsize=15)
plt.xlabel('')
plt.savefig('figure-1.png', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.ticker as mtick

pdf = df.loc[df['model'].str.startswith('mT5')].pivot('model', 'lang', 'pct') * 100
ax = pdf[[
        'English', 'Chinese', 'French', 'German',
        'Japanese', 'Russian', 'Spanish', 'Others',
    ]].plot(kind='bar', stacked=True, cmap=cm.get_cmap('Spectral'))
plt.legend(loc='upper right', bbox_to_anchor=(1.43, 0.95), fontsize=15)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(rotation=0, fontsize=15)
plt.yticks(rotation=0, fontsize=15)
plt.xlabel('')
plt.savefig('figure-2.png', bbox_inches='tight')
```
